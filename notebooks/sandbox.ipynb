### Графики для дашборда

```python
import plotly.graph_objects as go
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date']
```

```python
# Извлекаем год из даты для группировки
df['year'] = df['date'].dt.year
```

```python
# --- 2. Агрегация и расчет процентов ---
# Группируем данные по году и настроению, подсчитывая количество отзывов
sentiment_counts = df.groupby(['year', 'sentiment']).size().reset_index(name='counts')

# Преобразуем данные в удобный для графика формат (кросс-табуляция)
# и сразу рассчитываем проценты (normalize='index')
sentiment_pivot = pd.crosstab(
    index=sentiment_counts['year'],
    columns=sentiment_counts['sentiment'],
    values=sentiment_counts['counts'],
    aggfunc='sum'
).fillna(0)

# Рассчитываем процентное соотношение
sentiment_percentage = sentiment_pivot.div(sentiment_pivot.sum(axis=1), axis=0)

# --- 3. Визуализация с помощью Plotly ---

# Определяем цвета, похожие на ваш пример
# Вы можете изменить их на любые другие
# colors = {
#     'Negative': '#594138', # Темно-коричневый
#     'Neutral': '#E9E0D2',  # Светло-бежевый
#     'Positive': '#E9A431'  # Желто-оранжевый
# }

# Создаем объект фигуры
fig = go.Figure()

# Добавляем на график отдельный столбец (bar) для каждого типа настроения
for sentiment_type in sentiment_percentage.columns:
    fig.add_trace(go.Bar(
        x=sentiment_percentage.index,
        y=sentiment_percentage[sentiment_type],
        name=sentiment_type,
        # marker_color=colors.get(sentiment_type, '#CCCCCC'), # Используем серый цвет, если тип не найден
        text=sentiment_percentage[sentiment_type].apply(lambda x: f'{x:.0%}'), # Форматируем текст в проценты
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(size=14, color='white')
    ))

# --- 4. Настройка внешнего вида графика ---
fig.update_layout(
    barmode='stack', # Указываем, что это стековый график
    title_text='Распределение настроений по отзывам за разные годы',
    xaxis_title='Год',
    yaxis_title='Процентное соотношение',
    plot_bgcolor='white', # Белый фон
    xaxis=dict(
        showline=True,
        linewidth=1,
        linecolor='lightgrey',
        tickmode='array',
        tickvals=sentiment_percentage.index, # Явно указываем метки на оси X
        ticktext=[str(year) for year in sentiment_percentage.index]
    ),
    yaxis=dict(
        showgrid=False, # Убираем сетку по оси Y
        showticklabels=False, # Скрываем метки на оси Y, так как они есть на графике
        showline=False
    ),
    legend=dict(
        orientation='h', # Горизонтальная легенда
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1
    ),
    uniformtext_minsize=8,
    uniformtext_mode='hide' # Скрывать текст, если он не помещается
)

# Показываем график
fig.show()
```

```python
# --- 2. Фильтрация данных по дате ---
# Убедимся, что колонка 'date' имеет тип datetime (уже сделано при создании)
# df['date'] = pd.to_datetime(df['date']) # Эта строка нужна, если даты загружены как текст

# Определяем начальную и конечную дату для фильтрации
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2025-05-31')

# Применяем фильтр к DataFrame.
# .copy() используется, чтобы избежать предупреждения SettingWithCopyWarning от pandas.
df_filtered = df[(df['date'] >= start_date) & (df['date'] <= end_date)].copy()

# --- 3. Агрегация отфильтрованных данных ---
# Преобразуем дату в формат 'Год-Месяц' для группировки
# Важно: теперь мы работаем с df_filtered, а не с исходным df
df_filtered['year_month'] = df_filtered['date'].dt.strftime('%Y-%m')

# Группируем отфильтрованные данные и считаем количество
sentiment_counts = pd.crosstab(
    index=df_filtered['year_month'],
    columns=df_filtered['sentiment']
).fillna(0)


# --- 4. Визуализация с помощью Plotly ---

# Определяем цвета
# colors = {
#     'Negative': '#594138',
#     'Neutral': '#E9E0D2',
#     'Positive': '#E9A431'
# }

# Создаем объект фигуры
fig = go.Figure()

# Добавляем на график столбцы для каждого типа настроения
for sentiment_type in sentiment_counts.columns:
    fig.add_trace(go.Bar(
        x=sentiment_counts.index,
        y=sentiment_counts[sentiment_type],
        name=sentiment_type,
        # marker_color=colors.get(sentiment_type, '#CCCCCC'),
        text=sentiment_counts[sentiment_type],
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(size=14, color='white')
    ))

# --- 5. Настройка внешнего вида графика ---
fig.update_layout(
    barmode='stack',
    title_text='Абсолютное количество отзывов (с 01.01.2024 по 31.05.2025)',
    xaxis_title='Месяц',
    yaxis_title='Количество отзывов',
    plot_bgcolor='white',
    xaxis=dict(
        showline=True,
        linewidth=1,
        linecolor='lightgrey',
        type='category'
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='lightgray'
    ),
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1
    ),
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)

# Показываем график
fig.show()
```

```python
def identify_topic_by_subtopic(selected_topic, return_if_subtopic=True):
    selected_topic = selected_topic.strip()
    if selected_topic in topics_subtopics:
        return [selected_topic]
    
    topic_exists = False
    
    identified_subtopics = []
    for topic in topics_subtopics:
        subtopics = topics_subtopics[topic]
        if selected_topic in subtopics:
            topic_exists = True
            identified_subtopics.append(topic)
    
    if topic_exists and return_if_subtopic:
        identified_subtopics.append(selected_topic)
    
    return identified_subtopics
```
