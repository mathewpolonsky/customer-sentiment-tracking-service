```python
import pandas as pd
```

### Отзывы с банки.ру

```python
banki_dataset = pd.read_csv('../datasets/banki_ru/banki_dataset.csv')
banki_dataset['date'] = pd.to_datetime(banki_dataset['date'])
banki_dataset
```

```python
# Сортируем по дате
banki_dataset.sort_values(by='date', ascending=False, inplace=True)
banki_dataset
```

```python
# Переименовываем колонки
banki_dataset.rename(columns={'review_id': 'site_specific_id', 'service_name': 'source_topic'}, inplace=True)
banki_dataset
```

```python
# Добавляем источник
banki_dataset['source'] = 'banki.ru'
banki_dataset
```

```python
# Добавляем подтему из источника
banki_dataset['source_subtopic'] = None
banki_dataset
```

```python
# Переставляем колонки в нужном порядке
banki_dataset = banki_dataset.loc[:, [
    'site_specific_id', 'source', 'date', 'review_text', 'source_topic',
    'source_subtopic', 'rating'
]]
banki_dataset
```

```python
# Заменяем 'БЕЗ ОЦЕНКИ' на -1
banki_dataset.loc[banki_dataset['rating'] == 'БЕЗ ОЦЕНКИ', 'rating'] = -1
```

```python
# Нормализуем дату
banki_dataset['date'] = pd.to_datetime(banki_dataset['date']).dt.normalize()
banki_dataset
```

```python
# Ограничиваем даты до последней даты при парсинге сравни.ру
banki_dataset = banki_dataset[pd.to_datetime('2025-09-20') > banki_dataset['date']]
banki_dataset
```

### Отзывы со сравни.ру

```python
reviews_only_sravni_ru_gemma = pd.read_csv("../datasets/csv_for_db_structure/only_sravni_ru_gemma/reviews.csv")
reviews_only_sravni_ru_gemma['date'] = pd.to_datetime(reviews_only_sravni_ru_gemma['date'])
reviews_only_sravni_ru_gemma.sort_values(by='date', ascending=False, inplace=True)
reviews_only_sravni_ru_gemma
```

```python
# Убираем старую колонку reviewId
reviews_only_sravni_ru_gemma.drop('reviewId', axis=1, inplace=True)
reviews_only_sravni_ru_gemma
```

```python
# Переименовываем колонки
reviews_only_sravni_ru_gemma.rename(columns={
    'idSiteSpecific': 'site_specific_id', 'topic': 'source_topic',
    'subtopic': 'source_subtopic'}, inplace=True)
reviews_only_sravni_ru_gemma
```

```python
import numpy as np

reviews_only_sravni_ru_gemma = reviews_only_sravni_ru_gemma.replace(np.nan, None)
reviews_only_sravni_ru_gemma
```

---

### Объединение таблиц

```python
# --- Шаг 1: Объединение двух датафреймов ---
# Используем pd.concat для простого "склеивания" таблиц по вертикали.
# ignore_index=True здесь очень важен: он говорит pandas "забудь про старые
# индексы и создай новый, непрерывный индекс для объединенной таблицы".
# Это автоматически решает вашу проблему с неправильными индексами в banki_dataset.
reviews_df = pd.concat([banki_dataset, reviews_only_sravni_ru_gemma], ignore_index=True)
reviews_df
```

```python
# --- Шаг 2: Сортировка по дате ---
# kind='mergesort' - это стабильный алгоритм сортировки.
# "Стабильный" означает, что если несколько строк имеют одинаковое значение
# в колонке 'date', их относительный порядок (тот, что был после объединения)
# будет сохранен.
reviews_df = reviews_df.sort_values(by='date', ascending=False, kind='mergesort')
reviews_df
```

```python
# --- Шаг 3: Пересоздание индекса после сортировки ---
# После сортировки наш индекс теперь снова "сломан" (идет не по порядку).
# Используем reset_index(drop=True), чтобы создать новый, чистый,
# непрерывный индекс от 0 до N-1.
# drop=True говорит "не создавай новую колонку 'index' со старыми значениями".
reviews_df = reviews_df.reset_index(drop=True)
reviews_df
```

```python
# --- Шаг 4: Создание новой колонки 'id' ---
# Теперь, когда у нас есть идеальный финальный индекс, мы просто
# копируем его значения в новую колонку 'id'.
reviews_df['id'] = reviews_df.index

# --- Шаг 5: Установка 'id' как первой колонки ---
# Получаем список всех колонок, убираем из него 'id' и вставляем 'id' в начало.
# Затем используем этот новый список для переупорядочивания колонок.
cols = list(reviews_df.columns)
cols.remove('id')
reviews_df = reviews_df[['id'] + cols]
reviews_df
```

```python
# Сохраняем все спаршенные отзывы
reviews_df.to_csv('../datasets/csv_for_db_structure/all_parsed_reviews_db_structured/reviews.csv', index=False)
```
