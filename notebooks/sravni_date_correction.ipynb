```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('../datasets/sravni_ru/sravni_dataset.csv')
df
```

```python
# Словарь для замены русских месяцев (в родительном падеже)
month_map = {
    'января': '01', 'февраля': '02', 'марта': '03',
    'апреля': '04', 'мая': '05', 'июня': '06',
    'июля': '07', 'августа': '08', 'сентября': '09',
    'октября': '10', 'ноября': '11', 'декабря': '12'
}
```

```python
# Надежная функция для конвертации даты
def convert_date(date_str):
    # Заменяем название месяца на номер
    for month_ru, month_num in month_map.items():
        if month_ru in date_str:
            date_str = date_str.replace(month_ru, month_num)
            break
            
    # Добавляем текущий год, если в строке нет года
    if len(date_str.split()) == 2:
        date_str = f"{date_str} {datetime.now().year}"
        
    # Конвертируем в datetime, используя формат с числовым месяцем
    return pd.to_datetime(date_str, format='%d %m %Y')
```

```python
# Применяем функцию
df['date'] = df['date'].apply(convert_date)
```
