### Объединение таблиц и правка дат

```python
error_review_id = 12566223
main_reviews = main_reviews.drop(main_reviews[main_reviews['review_id'] == error_review_id].index)
reparsed_reviews = reparsed_reviews.drop(reparsed_reviews[reparsed_reviews['review_id'] == error_review_id].index)
```

```python
main_reviews.set_index('review_id', inplace=True)
reparsed_reviews.set_index('review_id', inplace=True)

main_reviews.update(reparsed_reviews)
main_reviews.reset_index(inplace=True)

main_reviews.iloc[48282:48282+5]
```

```python
main_reviews['date'] = pd.to_datetime(main_reviews['date'], format='%d.%m.%Y %H:%M')
main_reviews
```
